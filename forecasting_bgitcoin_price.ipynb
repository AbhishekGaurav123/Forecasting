```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()
    #Plot rolling statistics:
    plt.figure(figsize=(20,6))
    plt.plot(timeseries, color='yellow',label='Original')
    plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.show(block=False)
    print("Results of dickey fuller test")
    adft = adfuller(timeseries,autolag='AIC')
    # output for dft will give us without defining what the values are.
    #hence we manually write what values does it explains using a for loop
    output = pd.Series(adft[0:4],index=['Test Statistics','p-value','No. of lags used','Number of observations used'])
    for key,values in adft[4].items():
        output['critical value (%s)'%key] =  values
    print(output)

test_stationarity(dataset['market-price'])
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 10, 6
df_log = np.log(dataset['market-price'])
moving_avg = df_log.rolling(12).mean()
std_dev = df_log.rolling(12).std()
plt.legend(loc='best')
plt.title('Moving Average')
plt.plot(std_dev, color ="black", label = "Standard Deviation")
plt.plot(moving_avg, color="red", label = "Mean")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,6))
fig=plt.figure(1)
df_log_diff=df_log-df_log.shift()
plt.plot(df_log)
moving_avg = df_log.rolling(12).mean()
std_dev = df_log.rolling(12).std()
original=plt.plot(df_log_diff, color='yellow',label='Original')
mean=plt.plot(moving_avg, color='red', label='Rolling Mean')
rolling_std=plt.plot(std_dev, color='black', label = 'Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean and Standard Deviation')
plt.show(block=False)
```

```python
from statsmodels.tsa.stattools import acf, pacf
lag_acf=acf(df_log_diff, nlags=20)
lag_pacf=pacf(df_log_diff, nlags=20)
```

```python
import statsmodels.api as sm
fig=plt.figure(figsize=(16,6))
ax1=fig.add_subplot(211)
fig=sm.graphics.tsa.plot_acf(df_log_diff.dropna(),lags=40,ax=ax1)
ax2=fig.add_subplot(212)
fig=sm.graphics.tsa.plot_pacf(df_log_diff.dropna(),lags=40,ax=ax2)
```

```python
from statsmodels.tsa.arima_model import ARIMA as ar
plt.figure(figsize=(16,6))
model=ar(df_log,order=(4,1,5))
Arima_result=model.fit(disp=-1)
plt.plot(df_log_diff)
plt.plot(Arima_result.fittedvalues,color='yellow')
```

```python
Arima_diff_prediction=pd.Series(Arima_result.fittedvalues, copy=True)
print(Arima_diff_prediction.head())
```

```python
Arima_diff_prediction_cummulativesum=Arima_diff_prediction.cumsum()
print(Arima_diff_prediction_cummulativesum.head())
```

```python
Arima_log_prediction=pd.Series(df_log.iloc[0], index=df_log.index)
Arima_log_prediction=Arima_log_prediction.add(Arima_diff_prediction_cummulativesum,fill_value=0)
Arima_log_prediction.head()
```

```python
plt.figure(figsize=(16,6))
prediction=np.exp(Arima_log_prediction)
plt.plot(dataset['market-price'])
plt.plot(prediction)
plt.title('RMSE: %.4f'%np.sqrt(sum((prediction-dataset['market-price'])**2)/len(dataset['market-price'])))
```
